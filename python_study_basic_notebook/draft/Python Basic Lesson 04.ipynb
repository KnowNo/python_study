````python
#### 函数用法

函数是组织好的，可重复使用的，用来实现单一，或相关联功能的代码段。函数能提高应用的模块性，和代码的重复利用率。

python 提供了许多内建函数，比如print(), max()；

python 提供的各种功能的大量内建函数，可以满足绝大多数基本的功能

函数可以大大简化重复代码，降低工作量，减少错误

函数的基本语法

```
def 函数名称( 参数 ): 
	函数语句 
	return [值]
```

虽然 Python 内置了大量函数，但是大多数提供的是基本功能，有时候还是不能满足真正程序的需要
函数可以做的很复杂，发展为包(Package)。Python 目前有七万多个各类功能的包，几乎都可以可以免费使用在自己的程序中，这是最近几年 Python 飞速发展的原因之一。
学会使用函数来简化程序之后，可以为以后学习面向对象的编程方法打好基础。

r1 = 4
r2 = 6
r3 = 5.61
s1 = 3.14 * r1 * r1
s2 = 3.14 * r2 * r2
s3 = 3.14 * r3 * r3

print(s1)
print(s2)
print(s3)

def func1(r):
    s = 3.14 * r * r
    return s

print(func1(4))
print(func1(6))
print(func1(5.61))

max(1,5)

max(1,4,6,3,9)

# 请注意，函数体内部的语句在执行时，一旦执行到return时，函数就执行完毕，并将结果返回。
# 因此，函数内部通过条件判断和循环可以实现非常复杂的逻辑。

def my_max(a,b):
    if a > b:
        return a
    else:
        return b
    
print(my_max(3,6))

# 函数的不同参数形式--下节课再讲！
# *number理解为可以传入任意数目个参数，包括0个参数

def my_max(*number):
    print(number)
    return 100

print(my_max(1,5,30,14,25))

def my_max(*number):
    a_list = list(number)
    a_list.sort()
    a = a_list[-1]
    print(a_list)
    return a

print(my_max(1,5,30,14,25))

# Python的函数返回多值其实就是返回一个tuple，但语法上可以省略括号，写起来更方便

def my_max_min(*number):
    a_list = list(number)
    a_list.sort()
    my_max = a_list[-1]
    my_min = a_list[0]
    print(a_list)
    return my_max, my_min

print(my_max_min(1,5,30,14,25))    

---

* 写一个函数，输入一个数字，判断其是否是偶数，是偶数的话，返回 True，反之返回 False
* 写一个函数，输入一个英文单词，返回其中的元音字母的个数

a = int(input('Please input a number:'))

def oushu():
    if int(a/2) == a/2:
        return 2
        #print('这是个偶数')
    else:
        return 3
        #print('这是个奇数')
        
result = oushu()   
print(result)

a = input('plese input a word:')
b = ['a','e','i','o','u']
c = 0

for i in list(a):
    if i in b:
        c += 1
print(c)
````

````python
# 简单的剪刀石头布

import random

FIRST = 0
SECOND = 1 
BOTH = 2 

t1 = ('剪刀', '石头', '布')
t2 = ('human win', 'computer win', 'draw')

def which_win(i1, i2):
    if i1 == 0 and i2 == 1:
        return SECOND
    if i1 == 0 and i2 == 2:
        return FIRST
    if i1 == 1 and i2 == 0:
        return FIRST
    if i1 == 1 and i2 == 2:
        return SECOND
    if i1 == 2 and i2 == 0:
        return SECOND
    if i1 == 2 and i2 == 1:
        return FIRST
    if i1 == i2:
        return BOTH

print('0:剪刀 1:石头 2:布')
human = int(input('你出了:'))

c_index = random.randint(0,2)
computer = t1[c_index]

print("电脑出了",computer)

print(t2[which_win(human, c_index)])


思考：
* 如何优化改进上面的代码？
* 如何设计记录5局3胜的功能？
````
